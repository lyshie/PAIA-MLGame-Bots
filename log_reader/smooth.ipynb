```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = []
for j in ['maze.json']:
    data = pd.read_json(j)
    dfs.append(data)

dataset = pd.concat(dfs, ignore_index=True)
```

```python
fig = plt.figure(figsize=(16,4))

count = 500
plt.plot(dataset.index[:count], dataset['F_sensor'][:count], label='Raw')
plt.plot(dataset.index[:count], dataset['smooth_front'][:count], label='Smooth')

plt.xlabel('No.')
plt.ylabel('Distance')
plt.legend()
```
